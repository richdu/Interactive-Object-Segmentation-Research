```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sys, os
```

```python
DIR = '../../data/working/04-08-2019/'
ious = []
for fp in os.listdir(DIR):
    if "iou" in fp:
        ious.append(fp)
```

```python
iou_arrays = [np.load(os.path.join('../../data/working/04-08-2019/',fp)) for fp in ious]
iou_arrays = np.array(iou_arrays)
print(iou_arrays.shape)
```

```python
percentiles = np.percentile(iou_arrays, [25, 50, 75], axis=0)
average = np.mean(iou_arrays, axis=0)
```

```python
fig = plt.figure(figsize=(10,7))
plt.plot(percentiles[0], xlabel='25th percentile')
plt.plot(percentiles[1], label='50th percentile')
plt.plot(percentiles[2], label='75th percentile')
plt.plot(average, label='average')
plt.xticks(np.arange(21))
plt.title('IOU vs Iterations, N = 1000')
plt.xlabel('Iterations')
plt.ylabel('IOU')
plt.legend()
plt.show()
```
